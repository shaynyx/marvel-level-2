# **GBM**
 **Gradient Boosting Machine**, which is a type of machine learning algorithm belonging to the ensemble learning family. It is particularly popular for both classification and regression tasks. The core idea behind GBM is to build a series of weak learners (typically decision trees) sequentially, with each tree correcting the errors made by the previous ones.

image.png

```python
# Import necessary libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
# Create a Gradient Boosting classifier
gradient_boosting = GradientBoostingClassifier(n_estimators=50, learning_rate=0.05, random_state=42)


# Train the classifier on the training data
gradient_boosting.fit(X_train, y_train)
```

```python

# Make predictions on the testing data
y_pred = gradient_boosting.predict(X_test)
```

```python
# Calculate and print the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')

# Display additional metrics
print('\nClassification Report:')
print(classification_report(y_test, y_pred))
```
